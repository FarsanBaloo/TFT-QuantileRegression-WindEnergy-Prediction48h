```python
import optuna
import pandas as pd

# Ladda studien
study = optuna.load_study(study_name="tft_study", storage="sqlite:///tft_study.db")
df_trials = study.trials_dataframe()

# Filtrera bort försök utan resultat
df_trials = df_trials[df_trials["value"].notnull()]

# Sortera på value (lägsta = bäst)
df_sorted = df_trials.sort_values("value")

# Spara allt till CSV
df_sorted.to_csv("optuna_trials_log.csv", index=False)

panda = pd.read_csv("optuna_trials_log.csv")
panda.to_excel("optuna_trials_log.xlsx", index=False)
panda
```
